## Распределение Хи-квадрат

[Распределение Хи-квадрат](https://ru.wikipedia.org/wiki/Распределение_хи-квадрат) с _k_ степенями свободы — это распределение суммы квадратов _k_ независимых стандартных нормальных случайных величин.

**Обозначение:**

$\chi^2(k)$

**Определение:**

Пусть $z_1, \ldots, z_k$ — совместно независимые стандартные нормальные случайные величины, то есть:  $z_i \sim N(0,1) $. Тогда случайная величина

$x = z_1^2 + \cdots + z_k^2$

имеет распределение хи-квадрат с $k$ степенями свободы, то есть $x \sim f_{\chi^2(k)}(x)$.

**Плотность вероятности:**

![](https://upload.wikimedia.org/math/b/e/f/bef279b2cb4f25a855b38092b3dc671b.png)

**Функция распределения:**

![](https://upload.wikimedia.org/math/4/1/d/41ddcbb3e75722d31906cfc83b5902a9.png)

```python
# pls import
import numpy as np
import scipy.stats as sts
import matplotlib.pyplot as plt
%matplotlib inline
```

## Шаг 1 

Сгенерируем выборку объёма 1000 из $\chi^2_k$-распределения с заданным параметром $k$ = 5 (число степеней свободы):

```python
k = 5

rv = sts.chi2(k)
sample = rv.rvs(1000)
```

Построим гистограмму выборки и нарисуем поверх неё теоретическую плотность распределения нашей случайной величины.

```python
x = np.linspace(rv.ppf(0.01), rv.ppf(0.99), 100)
plt.hist(sample, bins=30, normed=True, alpha=0.5, label='histogram')
plt.plot(x, rv.pdf(x), color='r', lw=5, label='theoretical norm pdf')
plt.legend(frameon=False)
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.title("Histogram and theoretical PDF of $\chi^2_k$, $k=%s$" % k)
plt.show()
```

## Шаг 2

Далее мы для нескольких значений `n` (например, 5, 10, 50) сгенерируем 1000 выборок объёма `n` и построим гистограммы распределений их выборочных средних. Поверх каждой гистограммы нарисуем плотность соответствующего нормального распределения.

Чтобы это сделать, мы, используя информацию о среднем и дисперсии исходного распределения, посчитаем значения параметров нормальных распределений, которыми, согласно центральной предельной теореме, приближается распределение выборочных средних. Используем теоретические среднее и дисперсию (а не их выборочные оценки), которые выглядят следующим образом:

* $E(X) = k$ 
* $D(X) = 2k$

```python
mu = k
d = 2 * k 
# или, иначе:
sigma = np.sqrt(2 * k)
```

```python
size = 1000
ns = [5, 10, 50]

for n in ns:
    sample_means = [np.mean(rv.rvs(n)) for _ in range(1000)] # выборочные средние по 1000 выборкам объёма n
    
    # считаем параметры нормального распределения
    mu = k
    sigma = np.sqrt(2 * k)
    
    # задаём нормально распределённую случайную величину
    norm_rv = sts.norm(loc=mu, scale=sigma)
    
    # рисуем график
    x = np.linspace(norm_rv.ppf(0.001), norm_rv.ppf(0.999), 100)
    bins = int(50/np.sqrt(n))
    plt.hist(sample_means, bins=bins, alpha=0.5, normed=True, label='sample means')
    plt.plot(x, norm_rv.pdf(x), lw=4, color='red', label='theoretical PDF')
    plt.legend(frameon=False)
    plt.ylabel('$f(x)$')
    plt.xlabel('$x$')
    plt.title("CLT for $\chi^2_k$, $k=%s$ ($n=%s$ samples)" % (k, n))
    plt.show()
```

## Шаг 3

> Опишите разницу между полученными распределениями при различных значениях `n`. Как меняется точность аппроксимации распределения выборочных средних нормальным с ростом `n`?

Полученные распределения различаются, однако во всех случаях достаточно высока степень «похожести» гистограмм выборочных и графиков плотностей нормального распределения, и по внешнему виду данных гистограмм мы можем сделать предварительный вывод о том, что распределения выборочных средних принадлежат нормальному закону.

С ростом `n` исчезает асимметрия, уменьшается дисперсия распределения выборочных средних вокруг математического ожидания (теоретического среднего, которое в нашем случае равно $5$), а точность его аппроксимации нормальным распределением увеличивается. Как известно, при большом объеме выборки `n` в силу теоремы Чебышева среднее арифметическое значений случайных величин стремится по вероятности к математическому ожиданию каждой из них (иначе, к генеральному среднему). Таким образом, при дальнейшем росте выборки результат аппроксимации выборочных средних будет также расти.
