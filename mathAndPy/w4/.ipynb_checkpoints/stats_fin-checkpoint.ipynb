```python
%pylab inline
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
```

## Логарифмическое распределение
https://en.wikipedia.org/wiki/Logarithmic_distribution

```python
p = 0.1
rv = sts.logser(p)
samples = rv.rvs(1000)
```

```python
mu = -1/(log(1-p))*(p/(1-p))
var = -p*(p+log(1-p))/(((1-p)**2)*log(1-p)**2)
sigma = sqrt(var)
print 'Теоретическое среднее: %.4f'%mu
print 'Теоретическая дисперсия: %.4f'%var
```

```python
count, bins, ignored = plt.hist(samples, 20, normed=True, label='Logarithmic distribution samples hist')
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (bins - mu)**2 / (2 * sigma**2)), linewidth=2, color='r', label='Theoretical normal pdf')
plt.ylabel('Probability')
plt.xlabel('$x$')
plt.legend(loc='best', frameon=False)
plt.show()
```

```python
#Процедура для отображения гистограммы по выборкам
def showSmoothedHist(samples, sample_n):
    #Генерируем выборку средних по n
    sm_samples = [0.0]*1000
    d = dict()
    for i in xrange(len(sm_samples)):
        #генерируем выборки по n элементов
        sm_samples[i] = np.mean(rv.rvs(sample_n))
        #считаем кол-во уникальных значений (= кол-во столбиков гистограмм)
        d[sm_samples[i]] = 1.0
        
    #Считаем среднее по выборкам
    sm_mu = mean(sm_samples)
    #Считаем дисперсию по выборкам
    sm_dx = 0.0
    for i in xrange(len(sm_samples)):
        sm_dx += ((sm_samples[i]-sm_mu)**2)/len(sm_samples)
    #Среднеквадратичное отклонение
    sm_sigma = sqrt(sm_dx)

    print '\n\nРаспределение выборочных средних при n=%d'%sample_n

    zvar = var/sample_n
    zsigma = sqrt(zvar)
    
    count, bins, ignored = plt.hist(sm_samples, len(d.keys()), normed=True, label='n-samples hist')
    plt.plot(bins, 1/(zsigma * np.sqrt(2 * np.pi)) * np.exp( - (bins - mu)**2 / (2 * zsigma**2)), linewidth=2, color='r', label='theor norm pdf')
    #plt.legend()
    plt.ylabel('Probability')
    plt.xlabel('$x$')
    plt.legend(loc='best', frameon=False)
    plt.show()
    print 'теоретическая дисперсия по n-средних %.4f'%zvar
    print 'выборочное среднее %.4f (err=%.4f)'%(sm_mu,abs(mu-sm_mu))
    print 'выборочная дисперсия по n-средних %.4f (err=%.4f)'%(sm_dx,abs(zvar-sm_dx))
    print 'апроксимированая дисперсия исходного распределения %.4f (err=%.4f)'%(sm_dx*sample_n,abs(sm_dx*sample_n-var))
```

```python
#выводим гистограммы для различных n
for n in [2,3,5,10,20,100,200,300,500]:
    showSmoothedHist(samples, n)
```

## Вывод

По графикам видно, что при увеличении размера выборок (параметра n), графики точнее апроксимируют нормальное распределение. Ошибки между апроксимируемыми и теоретическими величинами можно объяснить малой выборкой (1000 элементов), но тем не менее весьма незначительными (разница в значениях ~0.002 при n=500)
