# Центральная предельная теорема своими руками

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
```

## Шаг 1. Выбираем распределение

Выберите ваше любимое непрерывное распределение. Сгенерируйте из него выборку объёма 1000, постройте гистограмму выборки и нарисуйте поверх неё теоретическую плотность распределения вашей случайной величины.

Рассмотрим распределение хи-квадрат с параметром df=3.

```python
# распределение хи-квадрат
chi2_rv = sts.chi2(df=3)

# выборка объема 1000
sample = chi2_rv.rvs(1000)
#plt.style.use('ggplot')
# гистограмма
plt.hist(sample, normed=True, bins=30, label='histogramm CHI2')
plt.ylabel('probability')
plt.xlabel('$x$')

# строим PDF
x = np.linspace(0, 15, 1500)
pdf = chi2_rv.pdf(x)
plt.plot(x, pdf, '-r', label='theoretical PDF', lw=2)
plt.legend(loc='best')
plt.show()
```

## Шаг 2. Генерируем выборочные средние и проверяем ЦПТ

Для нескольких значений n (например, 5, 10, 50) сгенерируйте 1000 выборок объёма n и постройте гистограммы распределений их выборочных средних. Используя информацию о среднем и дисперсии исходного распределения (её можно без труда найти в википедии), посчитайте значения параметров нормальных распределений, которыми, согласно центральной предельной теореме, приближается распределение выборочных средних. Обратите внимание: для подсчёта значений этих параметров нужно использовать именно теоретические среднее и дисперсию вашей случайной величины, а не их выборочные оценки. Поверх каждой гистограммы нарисуйте плотность соответствующего нормального распределения.

Для распределения хи-квадрат MX=df=3 и DX=2*df=6. Теоретически, по ЦПТ:
- для случая n = 5,   дисперсия нормального распределения DX/n = 6/5
- для случая n = 10,  дисперсия нормального распределения DX/n = 6/10
- для случая n = 50,  дисперсия нормального распределения DX/n = 6/50
- для случая n = 100, дисперсия нормального распределения DX/n = 6/100

```python
# 1000 выборок по 5
n = 5
sample = [np.mean(chi2_rv.rvs(n)) for i in range(1000)]

# строим гистограмму
plt.hist(sample, normed=True, bins=30, label='histogramm')
plt.ylabel('probability')
plt.xlabel('$x$')

# строим PDF нормального распределения
x = np.linspace(0, 8, 800)
norm_rv = sts.norm(loc=3, scale=np.sqrt(6./n))
plt.plot(x, norm_rv.pdf(x), '-r', label='normal PDF', lw=2)
plt.legend(loc='best')
plt.show()
```

### Вывод 1: для маленькой выборки можно наблюдать смещенную гистрограмму, так как распределение имеет асимметричный вид. И такого числа значений явно не достаточно.

```python
# 1000 выборок по 10
n = 10
sample = [np.mean(chi2_rv.rvs(n)) for i in range(1000)]

# строим гистограмму
plt.hist(sample, normed=True, bins=30, label='histogramm')
plt.ylabel('probability')
plt.xlabel('$x$')

# строим PDF нормального распределения
x = np.linspace(0, 6, 600)
norm_rv = sts.norm(loc=3, scale=np.sqrt(6./n))
plt.plot(x, norm_rv.pdf(x), '-r', label='normal PDF', lw=2)
plt.legend(loc='best')
plt.show()
```

### Вывод 2: при увеличении размера выборки в два раза, гистрограмма средних уже похожа на нормальное распределение.

```python
# 1000 выборок по 50
n = 50
sample = [np.mean(chi2_rv.rvs(n)) for i in range(1000)]

# строим гистограмму
plt.hist(sample, normed=True, bins=30, label='histogramm')
plt.ylabel('probability')
plt.xlabel('$x$')

# строим PDF нормального распределения
x = np.linspace(2, 4, 200)
norm_rv = sts.norm(loc=3, scale=np.sqrt(6./n))
plt.plot(x, norm_rv.pdf(x), '-r', label='normal PDF', lw=2)
plt.legend(loc='best')
plt.show()
```

### Вывод 3: при n=50 гистрограмма очень хорошо описывается соответствующим нормальным распределением.
