# Центральная предельная теорема своими руками

Выберите ваше любимое непрерывное распределение. Сгенерируйте из него выборку объёма 1000, постройте гистограмму выборки и нарисуйте поверх неё теоретическую плотность распределения вашей случайной величины.

**Выбрали распределение:** Бета-распределение;

Импортируем нужные библиотеки:

```python
from __future__ import division
from scipy.stats import beta,norm
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
plt.style.use('ggplot')
```

Задаем распределение, делаем выборку:

```python
a = 0.4
b = 0.5
bdistr = beta(a=a,b=b)
main_sample = bdistr.rvs(1000)
```

Строим гистограмму и теоретическую плотность распределения:

```python
x = np.linspace(0,1,100)
pdf = bdistr.pdf(x)
plt.plot(x,pdf)
plt.hist(main_sample, bins=20, normed=True)
plt.ylabel('$PDF(x)$')
plt.xlabel('$x$')
```

Для нескольких значений n (например, 5, 10, 50) сгенерируйте 1000 выборок объёма n и постройте гистограммы распределений их выборочных средних. Используя информацию о среднем и дисперсии исходного распределения, посчитайте значения параметров нормальных распределений, которыми, согласно центральной предельной теореме, приближается распределение выборочных средних. 

**Генерируем 1000 значений выборочных средних для нескольких значений n (5, 10, 50):**

```python
#Для выборок размером 5:
sampleMean5 = np.array([np.mean(np.random.choice(main_sample,5)) for i in\
                                                 xrange(1000)])

#Для выборок размером 10:
sampleMean10 = np.array([np.mean(np.random.choice(main_sample,10)) for i in\
                                                 xrange(1000)])    

#Для выборок размером 50:
sampleMean50 = np.array([np.mean(np.random.choice(main_sample,50)) for i in\
                                                 xrange(1000)])
```

Для применения ЦПТ нам нужно знать мат. ожидание и дисперсию исходного распределения.
Используя формулу из [Википедии](https://ru.wikipedia.org/wiki/%D0%91%D0%B5%D1%82%D0%B0-%D1%80%D0%B0%D1%81%D0%BF%D1%80%D0%B5%D0%B4%D0%B5%D0%BB%D0%B5%D0%BD%D0%B8%D0%B5) получаем:

$$\mu = \frac{\alpha}{\alpha+\beta} \\
\sigma^2 = \frac{\alpha  \beta}{(\alpha+\beta)^2(\alpha+\beta+1)} $$

```python
mu = a/(a+b)
disp = (a*b)/((a+b)**2 * (a+b+1))
```

Согласно ЦПТ, нормальное распределение, аппроксимирующее распределение выборочных средних задается формулой
$N(EX,\frac{DX}{n})$, где $EX$ - мат. ожидание, а $\frac{DX}{n}$ - дисперсия. 
Т.к. в scipy нормальное распределение задается через мат. ожидание и _среднеквадратичное отклонение_ (а не дисперсию), то в качестве параметра будем передавать $\sqrt{\frac{DX}{n}}$

Строим гистограмму для каждой выборки и плотность соответствующего нормального распределения.

Для распределения выборочных средних при размере выборки n=5:

```python
norm5 = norm(loc=mu, scale=np.sqrt(disp/5))
pdf5 = norm5.pdf(x)
plt.hist(sampleMean5, bins=20, normed=True)
plt.plot(x,pdf5)
plt.title('$N = 5$')
plt.ylabel('$PDF(x)$')
plt.xlabel('$x$')
```

Для распределения выборочных средних при размере выборки n=10:

```python
norm10 = norm(loc=mu, scale=np.sqrt(disp/10))
pdf10 = norm10.pdf(x)
plt.hist(sampleMean10, bins=20, normed=True)
plt.plot(x,pdf10)
plt.title('$N = 10$')
plt.ylabel('$PDF(x)$')
plt.xlabel('$x$')
```

Для распределения выборочных средних при размере выборки n=50:

```python
norm50 = norm(loc=mu, scale=np.sqrt(disp/50))
pdf50 = norm50.pdf(x)
plt.hist(sampleMean50, bins=20, normed=True)
plt.plot(x,pdf50)
plt.title('$N = 50$')
plt.ylabel('$PDF(x)$')
plt.xlabel('$x$')
```

### Выводы

Использование нормального распредлеления с параметрами выбранными по ЦПТ позволяет успешно аппроксимировать распределение выборочных средних, причем точность растет с ростом n (числа элементов для построения выборочного среднего). 

ЦПТ работает и вообще мимими.
